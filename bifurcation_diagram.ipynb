```python
from numpy.random import random as rng
from numpy import array, int16, arange, zeros
from matplotlib.pyplot import plot, show, scatter
```

```python
N = 1000
M = 1000

Lambda = arange(0, 1, 0.01)
X = zeros((Lambda.shape[0], N))
```

```python
def f(x, l):
  return 4*l*x*(1-x)
```

```python
for i in range(Lambda.shape[0]):
  for j in range(N):
    X[i][j] = rng()
    for k in range(M):
      X[i][j] = f(X[i][j], Lambda[i])
```

```python
print(X)
```

```python
for i in range(N):
  scatter(Lambda, X[:, i], color="black", s=2)
show()
```
